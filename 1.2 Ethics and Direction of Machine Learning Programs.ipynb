### Cleaning Data

```python
print(df.info())  # Quick overview
print(df.isnull().sum())  # Missing value count per column
```

```python
print(df.duplicated().sum())  # Count duplicates
df = df.drop_duplicates()     # Remove duplicates
```

#### No duplicate or missing values.

### Scaling Data

```python
# Separate date columns (keep them unchanged)
date_cols = ['DATE', 'MONTH']
df_dates = df[date_cols]

# Scale only the numeric feature columns
df_features = df.drop(columns=date_cols)
scaler = StandardScaler()
scaled_array = scaler.fit_transform(df_features)

# Reconstruct full DataFrame
scaled_df = pd.DataFrame(scaled_array, columns=df_features.columns)
final_df = pd.concat([df_dates.reset_index(drop=True), scaled_df], axis=1)

# Save scaled file using the defined path
save_path = os.path.join(path, '02 Data', 'Prepared Data', 'climate_data_scaled.csv')
final_df.to_csv(save_path, index=False)
```
